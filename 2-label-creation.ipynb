```python
import pandas as pd
features=pd.read_csv("ltv_features_raw.csv")

future_tx=pd.read_excel("future_transactions.xlsx")
```

```python
future_tx['order_date']=pd.to_datetime(future_tx['order_date'])
```

```python
future_tx=future_tx[(future_tx['order_date']>='2025-01-01') & (future_tx['order_date'] <='2025-12-31')]
future_tx.shape
```

```python
labels=future_tx.groupby('customer_id',as_index=False)['order_value'].sum()
labels=labels.rename(columns={'order_value':'ltv_12m'})
labels.head()
```

```python
df_train=features.merge(labels, on="customer_id",how='left')
df_train['ltv_12m']= df_train['ltv_12m'].fillna(0)
```

```python
df_train.to_csv("ltv_train.csv", index=False)
print("Saved ltv_train.csv --shape:",df_train.shape)
df_train.head()
```
